```python
# Importing python library
from sklearn.neighbors import NearestNeighbors
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
import numpy as np
import seaborn as sns
```

```python
# Read in csv file to each playlist and graph each attribute to spot differences. 
# I am reading in my happy and euphonious playlists.

pop = pd.read_csv("pop1.csv")
jazz = pd.read_csv("jazz1.csv")
rock = pd.read_csv("rock1.csv")
edm = pd.read_csv("edmy1.csv")


# plt.plot(mixed.energy, color='blue')
# plt.plot(mixed.valence, color='red')
# mixed['danceability']=mixed['danceability'].astype(int)
pop.isnull().values.any()
jazz.isnull().values.any()
rock.isnull().values.any()
edm.isnull().values.any()

plt.plot(pop.energy, color='blue')
plt.plot(jazz.energy, color='red')
plt.plot(rock.energy, color='green')
plt.plot(edm.energy, color='brown')

plt.xlabel("song #")
plt.ylabel("energy intensity")
plt.title("energy intensity across all songs in both playlists")
```

```python
plt.plot(pop.danceability, color='blue')
plt.plot(jazz.danceability, color='red')
plt.plot(rock.danceability, color='green')
plt.plot(edm.danceability, color='brown')

plt.xlabel("song #")
plt.ylabel("danceability intensity")
plt.title("danceability intensity across all songs in both playlists")
```

```python
plt.plot(pop.valence, color='blue')
plt.plot(jazz.valence, color='red')
plt.plot(rock.valence, color='green')
plt.plot(edm.valence, color='brown')

plt.xlabel("song #")
plt.ylabel("valence intensity")
plt.title("valence intensity across all songs in both playlists")
```
