### Fit the KNN Model on our data and use it to make prediction

```python
# Choosing Features
music_df = music_df.drop(["popularity", "duration_ms", "loudness"], axis=1)
```

```python
music_df.head(10)
```

**Shuffling dataset**

```python
np.random.seed(42)
sample_music = music_df.sample(frac=0.05)
sample_music.head()
```

```python
sample_music["target"].value_counts()
```

```python
len(sample_music)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(sample_music.drop('target', axis=1))

scaled_features = scaler.transform(sample_music.drop('target', axis=1))

music_df_features = pd.DataFrame(scaled_features, columns=sample_music.columns[:-1])
music_df_features
```

```python
from sklearn.model_selection import train_test_split

X = music_df_features
y = sample_music["target"]

X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state=30,
                                                    shuffle=True)
```

```python
import math
math.sqrt(len(y_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=23, p=2, metric='euclidean')
knn.fit(X_train, y_train)
```

```python
prediction = knn.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, prediction))
```
